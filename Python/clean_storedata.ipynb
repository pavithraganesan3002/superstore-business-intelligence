```python
df_cleaned=df.drop_duplicates(keep='first',inplace=True)
print(len(df))
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date']) 
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
df['Order_Day']=df['Order_Date'].dt.day
df['Order_Month']=df['Order_Date'].dt.month_name()
df['Order_Year']=df['Order_Date'].dt.year
df['Ship_Day']=df['Ship_Date'].dt.day
df['Ship_Month']=df['Ship_Date'].dt.month_name() 
df['Ship_Year']=df['Ship_Date'].dt.year
print(df.head())
```

```python
df = df.apply(lambda x: x.str.title() if x.dtype == "object" else x)
text_cols = ['Category', 'Sub_Category', 'City', 'State', 'Region', 'Segment']

for col in text_cols:
    df[col] = df[col].str.strip()
    
```

```python
df['Shipping_Delay'] = (df['Ship_Date'] - df['Order_Date']).dt.days
df['Profit_Margin'] = (df['Profit'] / df['Sales']).round(2) 
print(df.head())
```

```python
df['Sales'] = df['Sales'].round(2)
df['Profit'] = df['Profit'].round(2)
df['Discount'] = df['Discount'].round(2)
```

```python
df.to_csv("superstore_cleaned.csv", index=False)
print("Cleaned dataset saved!")
```
